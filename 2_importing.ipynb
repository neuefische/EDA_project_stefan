```python
# Connection to a PostgreSQL DB via psycopg2
import pandas as pd
import psycopg2
```

```python
import os
from dotenv import load_dotenv

load_dotenv()

DATABASE = os.getenv('DATABASE')
USER_DB = os.getenv('USER_DB')
PASSWORD = os.getenv('PASSWORD')
HOST = os.getenv('HOST')
PORT = os.getenv('PORT')
```

```python
# Create connection object conn
conn = psycopg2.connect(
    database=DATABASE,
    user=USER_DB,
    password=PASSWORD,
    host=HOST,
    port=PORT
)
```

```python
# Retrieving data from the database with psycopg2
# create cursor
cur = conn.cursor()

# run sql commands and fetch data
cur.execute('SELECT * FROM eda.king_county_house_sales')
cur.fetchall()
```

```python
# import the data into a pandas dataframe
query_string = "SELECT * FROM eda.king_county_house_sales"
df_psycopg = pd.read_sql(query_string, conn)
```

```python
#close the connection
conn.close()
```

```python
# show first 5 cases to check if importation worked
df_psycopg.head()
```

```python
#export the data to a csv-file
df_psycopg.to_csv('data/sales.csv',index=False)
```
